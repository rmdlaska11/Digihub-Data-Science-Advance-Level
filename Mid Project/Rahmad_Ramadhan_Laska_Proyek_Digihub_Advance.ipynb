## Uji Hipotesis

Case : Melakukan uji hipotesis menggunakan TTest untuk membuktikan apakah terdapat pengaruh antara terdapat atau tidaknya jumlah pemandangan air mempengaruhi harga rumah atau tidak.

Note : Disini, asumsi TTest bahwa data harus berdistribusi normal diabaikan karena kita belum mempelajari bagaimana caranya agar data berdistribusi normal.

```python
df.info()
```

### TTest

```python
df["waterfront"].unique()
```

```python
import scipy.stats as st

with_waterfront = df[df['waterfront'] == 1]
without_waterfront = df[df['waterfront'] == 0]

mean_with_waterfront = with_waterfront['price'].mean()
mean_without_waterfront = without_waterfront['price'].mean()
print("Mean price dengan waterfront:", mean_with_waterfront)
print("Mean price tanpa waterfront:", mean_without_waterfront)

ttest = st.ttest_ind(with_waterfront['price'], without_waterfront['price'])
p_value = ttest.pvalue
print("P-value:", p_value)

if p_value > 0.05:
    print("Terima H0, tidak cukup bukti mean price berbeda signifikan")
else:
    print("Terima H1, cukup bukti mean price berbeda signifikan")
```
